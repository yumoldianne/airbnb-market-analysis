```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Market Analysis EDA

```python
plt.hist(airbnb["openness"],bins=20)
plt.title("Distribution of Listings by Openness")
plt.show()
```

```python
sns.scatterplot(x="nightly_rate",y="length_stay", hue="guests", data=airbnb)
plt.title("Relationship between Nightly Rate and Average Length of Stay Categorized by Number of Guests")
```

```python
sns.scatterplot(x="nightly_rate",y="occupancy", hue ="guests", data=airbnb)
plt.title("Relationship between Nightly Rate and Occupancy Ratio Categorized by Number of Guests")
```

```python
sns.scatterplot(x="nightly_rate",y="occupancy", hue ="pool", data=airbnb)
plt.title("Relationship between Nightly Rate and Occupancy Ratio Categorized by Pool Availability")
```

```python
sns.scatterplot(x="nightly_rate",y="occupancy", hue ="hot_tub", data=airbnb)
plt.title("Relationship between Nightly Rate and Occupancy Ratio Categorized by Hot Tub Availability")
```

```python
revenue_total = airbnb.groupby(["city","host_type"])["revenue"].sum()
revenue_total = revenue_total.unstack(level=1)

revenue_total.plot(kind='bar', title="Total Revenue Generated by City and Host Type")
plt.show()
```

```python
revenue_total = airbnb.groupby(["city","bedrooms"])["revenue"].sum()
revenue_total = revenue_total.unstack(level=1)

revenue_total.plot(kind='bar', title="Total Revenue Generated by City and Number of Bedrooms")
plt.show()
```

```python
revenue_total = airbnb.groupby(["city","hot_tub"])["revenue"].sum()
revenue_total = revenue_total.unstack(level=1)

revenue_total.plot(kind='bar', title="Total Revenue Generated by City and Pool Availability")
plt.show()
```

```python
revenue_total = airbnb.groupby(["city","pool"])["revenue"].sum()
revenue_total = revenue_total.unstack(level=1)

revenue_total.plot(kind='bar', title="Total Revenue Generated by City and Hot Tub Availability")
plt.show()
```
